```python

cols = ['display_size', 'disk_space', 'discount_price', 'old_price', 'ratings_5max', 'graphics_card_score', 'processor_score']
df2[cols] = (df2[cols] - df2[cols].mean()) / df2[cols].std()
df2.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
df2_numpy = df2.values
similarity_matrix = cosine_similarity(df2_numpy)
sim_sorted = np.vstack([np.argsort(-row) for row in similarity_matrix])
np.save("similarity_mtrx", sim_sorted, allow_pickle=False)
```
